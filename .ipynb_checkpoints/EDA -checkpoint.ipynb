```python
dfc=pd.read_excel('Country-Code.xlsx')
dfc.head()
```

```python
dff=pd.merge(df,dfc,on='Country Code', how='left')#final dataframe containing the both datasets and left joining that
dff.head(2)
```

```python
dff.Country.value_counts()
#shows the number of times each country has come in the dataframe which tells  
#us the number of orders 
```

```python
names=dff.Country.value_counts().index
names
```

```python
chartvalues=dff.Country.value_counts().values
chartvalues
```

```python
plt.pie(chartvalues[:3],labels=names[:3])#top three countries by number of orders
```

```python
dff.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

observation:maximum transactions are from India,USA and United Kingdom in that order


```python
dff.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

this is the dataframe when we rename the column to rating count

```python
ratings=dff.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

Observations-

1.when rating is from 4.5 to 4.9---->Excellent i.e dark green

2.when rating is from 4.0 to 4.4---->very good i.e green

3.when rating is from 3.5 to 3.9----> good i.e yellow

4.when rating is from 2.5 to 3.4----> average i.e orange

5.when rating is from 1.8 to 2.4----> poor i.e red

6.when rating is 0---->not rated i.e white

```python
colours=ratings['Rating color'].unique()
colours
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['grey','red','orange','yellow','green','green'])
```

observations:-

1. There are a high number of no ratings 

2. Maximum number of ratings are from 2.5 to 3.4 that is average rating

the number of counties who have not rated
